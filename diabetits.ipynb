## Scaling

```python
df2=df.copy()
```

```python
from sklearn.preprocessing import RobustScaler
 
scaler = RobustScaler()
scaled_data = scaler.fit_transform(df2)
df3 = pd.DataFrame(scaled_data,columns=df.columns)
df3.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(df2)

# transform train and test sets
X_train_scaled = scaler.transform(df2)
scaled_df_stsc = pd.DataFrame(X_train_scaled,columns=df.columns)
scaled_df_stsc.head()
```

```python
from sklearn.preprocessing import MaxAbsScaler
max_abs=MaxAbsScaler()
scaled_data=max_abs.fit_transform(df2)
MaxAbsolute=pd.DataFrame(scaled_data,columns=df.columns)
MaxAbsolute.head()
```

```python
x=scaled_df_stsc.drop(columns='Outcome')
y=scaled_df_stsc['Outcome']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=32)
```

```python
from sklearn.svm import SVC

svc=SVC(kernel='rbf')

svc.fit(xtrain,ytrain)
# y_

# svc.coef_

## Prediction
y_pred=svc.predict(xtest)
# Evaluate the model
accuracy = accuracy_score(ytest, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
print("Classification Report:")
print(classification_report(ytest, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(ytest, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
Model=LogisticRegression()
Model.fit(xtrain,ytrain)# We fit our 80% of data into the model
p=Model.predict(xtest)
accuracy_score(p,ytest)*100
```
